```
import numpy as np
import scipy as sc
import matplotlib.pyplot as plt
```

```
from sklearn.datasets import load_boston
```

```
#cargamos libreria
boston = load_boston()

print(boston.DESCR)
```

Formula minimizar error cuadratico medio (MCO : $ \beta =  (X^{T}X)^{-1}X^TY$)

*   Elemento de lista
*   Elemento de lista



```
X = boston.data[:, 5]
Y = boston.target

#mostramos la grafica de x e y antes de añadir una dimension mas a la matriz x,
#necesaria ya que tenemos el termino independiente en nuestro metodo de minimos
#cuadrados
plt.scatter(X, Y, alpha=0.7)

print(X.shape)

#añadimos columnas de unos por el termino independiente
X = np.array([np.ones(506), X]).T
Y = np.array(Y)

print(X.shape)

B = np.linalg.inv(X.T @ X) @ X.T @ Y

#si colocamos unicamente una B entonces obtenemos el punto y donde corta y ademas
#tenemos su pendiente

plt.plot([4, 9], [B[0] + B[1]*4, B[0] + B[1]*9], c = "red")
plt.show()
```
